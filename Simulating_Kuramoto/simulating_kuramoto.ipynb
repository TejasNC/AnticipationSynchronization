## N = 2 

### Coupling Strength, K = 0.1

```python
model = KuramotoModel(N = 2, K = 0.1, T = 1000, dt = 0.1)
```

```python
model.run_simulation()
```

```python
print(model.find_critical_K())
```

So the current value of K is lower than the Critical Coupling Constant and therefore, there should be no coupling.

```python
model.plot_phases()
```

```python
model.plot_order_parameter()
```

```python
# Now, let's find the terminal value of the order parameter. For this, we will run the simulation for a longer time 
model = KuramotoModel(N=2, K=0.1, T=20000, dt=0.1)
model.run_simulation()
```

```python
model.find_terminal_order_parameter()
```

```python
# Let us plot the last 100 values of the order parameter
from matplotlib import pyplot as plt
plt.plot(model.R[-100:])
# plt.xlabel('Time')
plt.ylabel('Order Parameter')
plt.show()
```

Clearly, there is no coupling in this system.

### K = 2
As discussed earlier, this value of $K$ is above $K_C$

```python
model = KuramotoModel(N=100, K=2)
```

```python
model.run_simulation()
```

```python
model.plot_phases()
```

```python
# Plotting the phases for last 100 timesteps
from matplotlib import pyplot as plt
plt.plot(model.theta_t[-100:])
plt.ylabel("Phases")
plt.show()
```

Looks like there is some synchronization.

```python
model.plot_order_parameter()
```

```python
model.find_terminal_order_parameter()
```

So the order parameter is nearer to 1 consistently showing some degree of synchronization. So what is happening here is that that the population splits into 2 groups, with one group being coupled and the other group oscillating with their natural frequeicies. This is evident from the phase plot. As we increase K, the former group becomes larger and larger.
